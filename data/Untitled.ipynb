```python
import pandas as pd

def get_sample_gene_summary():
    sample_gene_summary = pd.read_csv('all_sample.gene_coverage.sample_gene_summary', delimiter='\t', index_col=False)
    return sample_gene_summary
```

```python
df = get_sample_gene_summary()
```

```python
df.columns
```

```python
def prepare_df(df):
    columns = df.columns.tolist()
    unique_columns = list(set(["_".join(col.split('_')[1:]) for col in columns[3:]]))
    
    for column in unique_columns:
        columns_for_mean = [original_column_name for original_column_name in df.columns if column in original_column_name]
        df[column] = df[columns_for_mean].mean(axis=1)
        
    return df.drop(columns=columns[3:])
```

```python
df2 = prepare_df(df)
```

```python
df2.columns
```

```python
df2.head()
```

```python
sample_gene_summary = df2
sample_gene_summary['gene_name'] = sample_gene_summary.Gene.apply(lambda x: x.split('_')[0])
df3 = sample_gene_summary.loc[sample_gene_summary['gene_name'].isin(['DAZ1', 'TNFRSF18'])]
```

```python
df4 = pd.DataFrame()
for x, y in df3.groupby('gene_name'):
    print(x)
    prepare_df(y)
```
